Now, lets perform above text processing on data.

```python
# Lower casing

zomato['reviews_list'] = zomato['reviews_list'].str.lower()

# Removal of punctuation

import string

PUNCT_TO_REMOVE = string.punctuation

print(PUNCT_TO_REMOVE)

def remove_punctuation(text):
    """Custom function to remove punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

zomato['reviews_list'] = zomato['reviews_list'].apply(lambda text:remove_punctuation(text))

```

```python
zomato
```

```python
# Removal of words

import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords

STOPWORDS = set(stopwords.words('english'))

def remove_stopwords(text):
    """Custom function to remove stopwords"""
    return ' '.join([word for word in str(text).split() if word not in STOPWORDS])

zomato['reviews_list'] = zomato['reviews_list'].apply(lambda text:remove_stopwords(text))

zomato
```

```python
# Removal of URLs

def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

zomato['reviews_list'] = zomato['reviews_list'].apply(lambda text:remove_urls(text))

zomato
```

# TF-IDF Vectorization

TF-IDF (Term frequency Inverse document frequency) vectors for each document. This will give you a matrix where each column represet a word in general in vocabulary (all words that appear in atleast one document) and each column represents a resturant, as before. 

TF-IDF is a statistical method of accessing the meaning of a word in a given document. Now I will use TF-IDF vectorization on dataset. 

```python
df_percent = zomato.sample(frac=0.5)
```

```python
df_percent.set_index('name', inplace=True)
```

```python
df_percent
```

```python
indices = pd.Series(df_percent.index)
```

```python
indices
```

```python
# Creating TF-IDF matrix

tfidf = TfidfVectorizer(analyzer='word', ngram_range=(1,2), min_df=0, stop_words='english')

tfidf_matrix = tfidf.fit_transform(df_percent['reviews_list'])
print(tfidf_matrix)
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
print(cosine_similarities)
```

Now, last step of creating resturant recommendation system is to write a function which will do resturant recommendation.

```python
def recommend(name, cosine_similarities=cosine_similarities):
    
    # Create a list to put top resturants
    recommend_resturant = []
    
    # Find the index of the hotel entered
    idx = indices[indices==name].index[0]
    
    #Find the resturants with similar cosine similarity value and order them
    score_series = pd.Series(cosine_similarities[idx]).sort_values(ascending=False)
    
    #Extract top 30 resturant indexes with a similar cosine sim value
    top30_index = list(score_series.iloc[0:31].index)
    
    # Names of the top 30 resturant
    for i in top30_index:
        recommend_resturant.append(list(df_percent.index)[i])
        
    # Create the new dataset to show similar resturants
    df_new = pd.DataFrame(columns=['cuisines', 'mean_rating', 'cost'])
    
    # Create top 30 similar resturants with some of their columns
    for i in recommend_resturant:
        df_new = df_new.append(pd.DataFrame(df_percent[['cuisines', 'mean_rating', 'cost']][df_percent.index==i].sample()))
        
    # Drop the same name resturant and sort only the top 10 by the highest rating
    df_new = df_new.drop_duplicates(subset=['cuisines', 'mean_rating', 'cost'], keep=False)
    df_new = df_new.sort_values(by='mean_rating', ascending=False)
    print('TOP %s resturants like %s with similar reviews: ' %(str(len(df_new)), name))
    
    return df_new
    
```
